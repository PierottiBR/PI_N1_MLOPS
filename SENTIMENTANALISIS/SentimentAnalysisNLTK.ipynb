```python
tokens = nltk.word_tokenize(example)
tokens[:10]
```

```python
tagged = nltk.pos_tag(tokens)
tagged[:10]
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('Great game!')
```

```python
sia.polarity_scores('Worst game ever!')
```

```python
sia.polarity_scores(example)
```

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
  text = row['review']
  res[i] = sia.polarity_scores(text)
```

```python
vaders = pd.DataFrame(res).T
vaders
```

```python
final_df = df.merge(vaders, left_index=True, right_index=True)
final_df
```

```python
final_df['sentiment_analysis'] = None
```

```python
def calcular_sentimiento(fila):
  valor1 = fila['neg']
  valor2 = fila['neu']
  valor3 = fila['pos']

  if valor3 > valor2 and valor3 > valor1:
    maximo = 2
  elif valor2 > valor1 and valor2 > valor3:
    maximo = 1
  else:
    maximo = 0


  return(maximo)
```

```python
final_df['sentiment_analysis'] = final_df.apply(calcular_sentimiento, axis=1)
```

```python
final_df
```

```python
df = final_df.drop(columns=['neg', 'neu', 'pos', 'compound'])
df = df.reindex(columns=['user_id', 'item_id', 'recommend', 'sentiment_analysis', 'posted','review'])
```
