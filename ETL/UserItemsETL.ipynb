```python
import pandas as pd
import ast
```

```python
rows = []

with open(r'D:\PROGRAMACION\PROYECTOS\PROYECTOINDIVIDIAL1\PROYECTOINDIVUDUAL\DATASETS\australian_users_items.json', encoding = 'utf-8') as f:
  for line in f.readlines():
    rows.append(ast.literal_eval(line))

df_items = pd.DataFrame(rows)
```

```python
df_items = pd.json_normalize(rows, record_path='items', meta=['user_id', 'user_url', 'items_count', 'steam_id'])
```

```python
df_items = df_items.drop(columns=['user_url', 'items_count', 'steam_id', 'playtime_2weeks'])
```

```python
df_items.drop_duplicates(subset=['user_id','item_id','item_name'],inplace=True)
```

```python
df_items
```

```python
df = df_items[df_items['playtime_forever'] != 0]
```

```python
df
```

```python
df = df.reindex(columns=['user_id', 'item_id', 'item_name', 'playtime_forever'])
```
